# Visualize Simulated Paths

Import modules

```python
import matplotlib.pyplot as plt
import numpy as np
import sys
from typing import List

sys.path.append("..")

from path import Path
from path_generator import generate_paths, Random, RandomStepRange
```

## Step Function: `Random`

```python
step_fn = Random(randomness=RandomStepRange(r_min=-5, r_max=5, d_min=5, d_max=20))

paths: List[Path] = generate_paths(sample_size=10, iterations=100, step_fn=step_fn)  #, initial_state={"r": -1.5})


canvas = np.zeros((1080, 1920, 3), np.uint8)

for path in paths:
    path.plot(canvas, thickness=5)

plt.imshow(canvas)
plt.show()
plt.close()
```
